```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vectors
v1 = np.array([1, 1])
v2 = np.array([-1, -1])
```

```python
# Graph vector lineal combinations a * v1 + b * v2 (for -10 <= a, b < 10 and a, b integers)
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v1[0] * a + v2[0] * b, v1[1] * a + v2[1] * b,
                   marker='.', color='blue')
        
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

```python
v3 = np.array([1, 0])
v4 = np.array([2, -3])
```

```python
# Graph vector lineal combinations a * v3 + b * v4 (for -10 <= a, b < 10 and a, b integers)
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v3[0] * a + v4[0] * b, v3[1] * a + v4[1] * b,
                   marker='.', color='blue')
        
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

v5 = np.array([1, 0, 0])
v6 = np.array([2, -3, 0])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for a in range(-10, 10):
    for b in range(-10, 10):
        ax.scatter(v5[0] * a + v6[0] * b, v5[1] * a + v6[1] * b, v5[2] * a + v6[2] * b,
                   marker='.', color='blue')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```
